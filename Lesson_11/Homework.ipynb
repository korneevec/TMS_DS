```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("files/img_1.jpg")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
kernel = np.array([[1, 2, 1],
                   [2, 4, 2],
                   [1, 2, 1]]) / 16.0
```

```python

def blur_image(kernel, img):
    img_h, img_w, img_c = img.shape
    
    copy_image = cv2.copyMakeBorder(img, 1, 1, 1, 1, cv2.BORDER_REPLICATE)
    
    for i in range(1, img_h + 1):
        for j in range(1, img_w + 1):
            for c in range(img_c):
                region = copy_image[i-1:i+2, j-1:j+2, c]
                img[i-1, j-1, c] = np.sum(kernel * region)
        
    return img
```

```python
img = blur_image(kernel, img)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
